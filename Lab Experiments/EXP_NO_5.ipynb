```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("banknotes.csv")

print(data.head(10))
```

```python
print(data.shape)

print(data.describe())
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]

print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
mlp = MLPClassifier(max_iter=500, activation='relu')


mlp.fit(x_train, y_train)


pred = mlp.predict(x_test)


print("Confusion Matrix (ReLU):")
print(confusion_matrix(y_test, pred))

print("Classification Report (ReLU):")
print(classification_report(y_test, pred))
```

```python
mlp = MLPClassifier(max_iter=500, activation='logistic')
mlp.fit(x_train, y_train)

pred = mlp.predict(x_test)

print("Confusion Matrix (Logistic):")
print(confusion_matrix(y_test, pred))

print("Classification Report (Logistic):")
print(classification_report(y_test, pred))
```

```python
mlp = MLPClassifier(max_iter=500, activation='logistic')
mlp.fit(x_train, y_train)

pred = mlp.predict(x_test)

print("Confusion Matrix (Logistic):")
print(confusion_matrix(y_test, pred))

print("Classification Report (Logistic):")
print(classification_report(y_test, pred))
```

```python
mlp = MLPClassifier(max_iter=500, activation='tanh')
mlp.fit(x_train, y_train)

pred = mlp.predict(x_test)

print("Confusion Matrix (Tanh):")
print(confusion_matrix(y_test, pred))

print("Classification Report (Tanh):")
print(classification_report(y_test, pred))
```

```python
mlp = MLPClassifier(max_iter=500, activation='identity')
mlp.fit(x_train, y_train)

pred = mlp.predict(x_test)

print("Confusion Matrix (Identity):")
print(confusion_matrix(y_test, pred))

print("Classification Report (Identity):")
print(classification_report(y_test, pred))
```
