```python
import numpy as np

X = np.array([[0,0],[0,1],[1,1],[1,0]])
y = np.array([[0],[0],[1],[0]])

W = np.array([[0.1],[0.3]])
b = 0.2

def sigmoid(x):
    return 1/(1+np.exp(-x))

def der(x):
    return sigmoid(x)*(1-sigmoid(x))

for i in range(15000):
    z = np.dot(X,W) + b
    out = sigmoid(z)
    error = out - y
    d = error * der(out)
    W = W - 0.05 * np.dot(X.T,d)
    b = b - 0.05 * np.sum(d)

print("Weights:",W)
print("Bias:",b)

test = np.array([[1,0],[1,1],[0,0],[0,1]])
print(sigmoid(np.dot(test,W)+b))
```
